```python
def update_user(user):
    response = requests.put(
        f"{sdw2023_api_url}/users/{user['id']}",
        json=user
    )
    return response.status_code == 200

for user in users:
    if user is None:
        print("Skipping update for a None user.")
        continue
    print(f"Atualizando {user['name']}...")
    print("OK!" if update_user(user) else "Falhou!")
```

```python
import requests
from openai import OpenAI
client = OpenAI(api_key="COLE_SUA_API_KEY_AQUI")

def generate_news(user):
    prompt = f"Crie uma mensagem de marketing curta incentivando {user['name']} a investir melhor seu dinheiro."

    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    return completion.choices[0].message.content

# Adicionando mensagens personalizadas
for user in users:
    if user is None:
        print("Skipping a None user as API call failed previously.")
        continue
    new_message = {
        "id": 999,
        "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/profit.svg",
        "description": generate_news(user)
    }
    if "news" not in user:
        user["news"] = []
    user["news"].append(new_message)

print("Mensagens geradas!")
```
